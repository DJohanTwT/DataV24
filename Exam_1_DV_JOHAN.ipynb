##**1.1. Variable Description**

<font face = "Times New Roman" color = "black" size = 4> There are the variables that I will use. </font>

* <font face = "Times New Roman" color = "red" size = 4> Exports of goods and services (% of GDP): </font> <font face = "Times New Roman" size = 4> Exports of goods and services represent the value of all goods and other market services provided to the rest of the world as a percentage of GDP. They include the value of merchandise, freight, insurance, transport, travel, royalties, license fees, and other services, such as communication, construction, financial, information, business, personal, and government services. They exclude compensation of employees and investment income (formerly called factor services) and transfer payments. </font>

* <font face = "Times New Roman" color = "red" size = 4> Imports of goods and services (% of GDP): </font> <font face = "Times New Roman" size = 4> Imports of goods and services represent the value of all goods and other market services received from the rest of the world as a percentage of GDP. They include and exclude the same services of exports. </font>

* <font face = "Times New Roman" color = "red" size = 4> Foreign direct investment, net inflows (% of GDP):</font> <font face = "Times New Roman" size = 4> Foreign direct investment are the net inflows of investment to acquire a lasting management interest in an enterprise operating in an economy other than that of the investor. It is the sum of equity capital, reinvestment of earnings, other long-term capital, and short-term capital. This series shows net inflows in the reporting economy from foreign investors, and it is expressed as a percentage of GDP.</font>


* <font face = "Times New Roman" color = "red" size = 4> Trade (% of GDP):</font> <font face = "Times New Roman" size = 4> Trade is the sum of exports and imports of goods and services measured as a share of gross domestic product, and it is expressed as a percentage of GDP. </font>


* <font face = "Times New Roman" color = "red" size = 4> GDP growth (annual %):</font>  <font face = "Times New Roman" size = 4> Annual percentage growth rate of GDP at market prices based on constant local currency. Aggregates are based on constant 2015 prices, expressed in U.S. dollars. GDP is the sum of gross value added by all resident producers in the economy plus any product taxes and minus any subsidies not included in the value of the products. It is calculated without making deductions for depreciation of fabricated assets or for depletion and degradation of natural resources. It is expressed as a percentage.  </font>


* <font face = "Times New Roman" color = "red" size = 4> External balance on goods and services (% of GDP):</font> <font face = "Times New Roman" size = 4> It's the trade balance and it equals exports of goods and services minus imports of goods and services, and it is expressed as a percentage of GDP.</font>
